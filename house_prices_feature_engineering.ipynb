# House Prices - Advanced Regression Techniques - Feature Engineering
We will be performing all the below steps in Feature Engineering-

1. Missing values
2. Temporal variables
3. Categorical variables: remove rare labels
4. Standarise the values of the variables to the same range

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# to visualise all the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
# Importing train and test csv files
dataset_train = pd.read_csv('train.csv')
dataset_test = pd.read_csv('test.csv')
```

## Missing values

```python
# Let us capture all the nan values
# First lets handle Categorical features which are missing
def missing_values(dataset):
    features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1 and dataset[feature].dtypes=='O']

    for feature in features_nan:
        print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean()*100,4)))
    
    return features_nan
```

```python
train_features_nan = missing_values(dataset_train)
```

```python
test_features_nan = missing_values(dataset_test)
```

```python
# Replace missing value with a new label
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

dataset_train = replace_cat_feature(dataset_train, train_features_nan)
dataset_test = replace_cat_feature(dataset_test, test_features_nan)
```

```python
dataset_train[train_features_nan].isnull().sum()
```

```python
dataset_test[test_features_nan].isnull().sum()
```

```python
# Now lets check for numerical variables that contains missing values
def missing_values_numerical(dataset):
    numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1 and dataset[feature].dtypes!='O']

    # We will print the numerical nan variables and percentage of missing values
    for feature in numerical_with_nan:
        print("{}: {}% missing value".format(feature,np.around(dataset[feature].isnull().mean()*100,4)))
        
    return numerical_with_nan
```

```python
train_num_features_nan = missing_values_numerical(dataset_train)
```

```python
test_num_features_nan = missing_values_numerical(dataset_test)
```

```python
# Replacing the numerical Missing Values
def replace_num_feature(dataset,numerical_with_nan):
    for feature in numerical_with_nan:
        # We will replace by using median since there are outliers
        median_value = dataset[feature].median()
    
        # create a new feature to capture nan values
        dataset[feature+'nan'] = np.where(dataset[feature].isnull(),1,0)
        dataset[feature].fillna(median_value,inplace=True)
    
    return dataset

dataset_train = replace_num_feature(dataset_train, train_num_features_nan)
dataset_test = replace_num_feature(dataset_test, test_num_features_nan)
```

```python
dataset_train[train_num_features_nan].isnull().sum()
```

```python
dataset_test[test_num_features_nan].isnull().sum()
```

```python
# Date Time Variables
def date_time_var(dataset):
    for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:  
        dataset[feature]=dataset['YrSold']-dataset[feature]
    return dataset

dataset_train = date_time_var(dataset_train)
dataset_test = date_time_var(dataset_test)
```

## Handling Rare Categorical Feature
We will remove categorical variables that are present less than 1% of the observations.

```python
def cat_features(dataset):
    categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
    return categorical_features
```

```python
categorical_features = cat_features(dataset_train)
for feature in categorical_features:
    temp=dataset_train.groupby([feature])[feature].count()/len(dataset_train)
    temp_df=temp[temp>0.01].index
    dataset_train[feature]=np.where(dataset_train[feature].isin(temp_df),dataset_train[feature],'Rare_var')
    dataset_test[feature]=np.where(dataset_test[feature].isin(temp_df),dataset_test[feature],'Rare_var')
```

Label Encoding to the categorical feature.

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

categorical_features = cat_features(dataset_train)
for feature in categorical_features:
    label_encoder.fit(pd.concat([dataset_train[feature], dataset_test[feature]]))
    dataset_train[feature] = label_encoder.transform(dataset_train[feature])
    dataset_test[feature] = label_encoder.transform(dataset_test[feature])
```

## Feature Scaling

```python
# Remove all the extra features from test dataset that are not in train dataset
dataset_test_new = pd.DataFrame()
train_features = dataset_train.drop(['SalePrice'],axis=1).columns
for feature in train_features:
    dataset_test_new[feature] = dataset_test[feature]
    
# Selecting features for feature scaling
feature_scale_train=[feature for feature in dataset_train.columns if feature not in ['Id','SalePrice']]
feature_scale_test=[feature for feature in dataset_test_new.columns if feature not in ['Id']]

# Performing feature scaling using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(pd.concat([dataset_train[feature_scale_train], dataset_test_new[feature_scale_test]]))
```

```python
# transform the train dataset, and add Id and SalePrice variables
mod_data_train = pd.concat([dataset_train[['Id', 'SalePrice']].reset_index(drop=True),
                           pd.DataFrame(scaler.transform(dataset_train[feature_scale_train]), columns=feature_scale_train)],
                           axis=1)

# transform the test dataset, and add Id variable
mod_data_test = pd.concat([dataset_test_new[['Id']].reset_index(drop=True),
                          pd.DataFrame(scaler.transform(dataset_test_new[feature_scale_test]), columns=feature_scale_test)],
                          axis=1)
```
